```python
import pandas as pd
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")

star_wars.head(10)
```

```python
print(star_wars.columns)
```

```python
star_wars = star_wars[star_wars['RespondentID'].notnull()]

star_wars.head(10)
```

So far we've imported our dataset and removed rows where RespondentID is null. 

Now we'll clean columns for analysis, starting with boolean values for yes/no questions.

```python
yes_no = {'Yes': True, 'No' : False}

star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] =  star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(yes_no)
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] = star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(yes_no)

star_wars.head()
```

Next up, boolean values for which movies participants have seen:

```python
import numpy as np 

movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movie_mapping)
    
star_wars.head()
```

Renaming columns to better reflect their meaning:

```python
name_mapping = {
    "Which of the following Star Wars films have you seen? Please select all that apply.": 'seen_1',
    "Unnamed: 4": 'seen_2',
    "Unnamed: 5": 'seen_3',
    "Unnamed: 6": 'seen_4',
    "Unnamed: 7": 'seen_5',
    "Unnamed: 8": 'seen_6'
}

star_wars = star_wars.rename(columns = name_mapping)

star_wars.head()
```

Renaming columns that rank the movies from best to worst, and converting their values to floats for analysis:

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)

ranking_dict = {'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.' : 'ranking_1', 'Unnamed: 10' : 'ranking_2', 'Unnamed: 11' : 'ranking_3', 'Unnamed: 12' : 'ranking_4', 'Unnamed: 13':'ranking_5', 'Unnamed: 14' : 'ranking_6'}
star_wars = star_wars.rename(columns = ranking_dict)

star_wars.head()
```

Plotting a chart of the mean rankings:

```python
%matplotlib inline
import matplotlib.pyplot as plt
star_wars[star_wars.columns[9:15]].mean()
plt.bar(range(6), star_wars[star_wars.columns[9:15]].mean())
```

Let's remember that the lowest ranking is the most popular, meaning the 5th movie (Empire Strikes Back) is on average the top ranked, followed by Return of the Jedi and A New Hope. On average, the first three Star Wars movies (4, 5, 6) are all higher ranked than any of the later releases (1, 2, 3). 

Next, we'll examine how many participants have seen each movie:

```python
star_wars[star_wars.columns[3:8]].sum()
plt.bar(range(6), star_wars[star_wars.columns[3:9]].sum())
```

People have mostly seen the 5th and 6th titles, which are also the highest ranked. However, the correlation drops off for the rest of the titles. 

Let's see how self-identified fans responded versus non-fans:

```python
fans = star_wars[star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] == True]
nonfans = star_wars[star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] == False]

plt.bar(range(6), fans[fans.columns[9:15]].mean())
plt.show()
plt.bar(range(6), nonfans[nonfans.columns[9:15]].mean())
```

```python
plt.bar(range(6), fans[fans.columns[3:9]].sum())
plt.show()
plt.bar(range(6), nonfans[nonfans.columns[3:9]].sum())
```

It looks like fans and nonfans have consistent views about the ranks of the films, with fans having slightly more polarized views on the rankings. 

Unsurprisingly, fans have reported seeing almost all of the films consistently, while nonfans were more likely to have seen the three most popular titles. 

Let's run the same analysis for males versus females:

```python
males = star_wars[star_wars['Gender'] == 'Male']
females = star_wars[star_wars['Gender'] == 'Female']

plt.bar(range(6), males[males.columns[9:15]].mean())
plt.show()
plt.bar(range(6), females[females.columns[9:15]].mean())
plt.show()
```

```python
plt.bar(range(6), males[males.columns[3:9]].sum())
plt.show()
plt.bar(range(6), females[females.columns[3:9]].sum())
plt.show()
```

Consistent rankings between men and women, with more nuanced ranking among the men. 

Viewership also follows similar trends, with males seeing the movies more than females.
